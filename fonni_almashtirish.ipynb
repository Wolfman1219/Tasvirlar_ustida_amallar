```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2

rasm = cv2.imread('/content/drive/MyDrive/man in green background.jpg')
rasm = cv2.cvtColor(rasm, cv2.COLOR_BGR2RGB)
plt.imshow(rasm)
plt.show()

rasm2 = np.copy(rasm)
```

```python
rasm2.shape
```

```python
kok = np.array([0, 230, 0])
kok2 = np.array([250, 255, 250])
maska = cv2.inRange(rasm2, kok, kok2)
plt.imshow(maska, cmap = 'gray')
plt.show()
```

```python
rasm2[maska != 0] = [0, 0, 0]
plt.imshow(rasm2) 
```

```python
fon = cv2.imread('/content/drive/MyDrive/dexus-office-space.jpg')
fon = cv2.cvtColor(fon, cv2.COLOR_BGR2RGB)
fon = fon[:338, :600]
fon[maska == 0] = [0, 0, 0]
tayyor = fon + rasm2
plt.imshow(tayyor)
plt.show()
```
